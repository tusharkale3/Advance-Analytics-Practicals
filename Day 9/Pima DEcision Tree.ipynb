```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.preprocessing import LabelEncoder 
from sklearn import tree 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
```

```python
col_names =['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label']
```

```python
pima = pd.read_csv(r"D:\DBDA\Adavance Stats\Day 9\pima-indians-diabetes.csv",header=None,names=col_names)

pima
```

```python
# split dataset into features and target variable

features_col =['pregnant','glucose','bp','insulin','bmi','pedigree','age']

X= pima[features_col]
y=pima.label 
```

```python
# Split dataset into training and testing set

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)

model = DecisionTreeClassifier()
model.fit(X_train,y_train)

y_pred =model.predict(X_test)

print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
from IPython.display import Image
import pydotplus
dot_data=StringIO()


export_graphviz(model,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=features_col,class_names=['0','1'])

graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('diabetes.png')

Image(graph.create_png())
```
