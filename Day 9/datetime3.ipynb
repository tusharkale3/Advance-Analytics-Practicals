```python
df = pd.read_csv("D:\DBDA\Adavance Stats\Day 9\starbucks.csv", index_col='Date', parse_dates=True)
print(df.head())
print(df.index)
```

```python
# Take year end frequency sample

print(df.resample(rule='A').mean())
```

```python
#visualize

import matplotlib.pyplot as plt 
df.Close.resample('A').mean().plot.bar()
plt.show()
```

```python
title ="Monthly closing price for Starbucks"

df.Close.resample('M').max().plot.bar(figsize=(16,6),title=title)
plt.show()
```

```python
# Check data range
print(df.head())
print(df.tail())

# Let us shift data by one row
print(df.shift(-5))
# Notice that original price of 2 Jan is now shown on 5 Jan, etc...
print(df.shift(-2).tail())
```

```python
print(df.shift(periods=1, freq='M').head())
```

```python
# Now let's add in a rolling mean This rolling method provides row entries, where every entry is then representaive of the window.

# 7 day rolling mean 
print(df.rolling(window=7).mean().head(15))

df['Close'].plot(figsize=(12,5))
df.rolling(window=7).mean()['Close'].plot();
plt.show()
```

```python
# 7 day rolling mean
pd.set_option('display.float_format', lambda x: '%.2f' % x) # Set to 2 decimal places
print(df.rolling(window=7).mean().head(15))

df['Close'].plot(figsize=(12,5))
df.rolling(window=30).mean()['Close'].plot();
plt.show()


```
