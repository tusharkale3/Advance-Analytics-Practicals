```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
odi_20 = pd.read_csv('G:\\gb\\Advance Analytics\\Theory\\Day_6\\Men ODI Player Innings Stats - 20th Century.csv')
odi_21 = pd.read_csv('G:\\gb\\Advance Analytics\\Theory\\Day_6\\Men ODI Player Innings Stats - 21st Century.csv')

df = pd.concat([odi_20,odi_21],axis=0)
df
```

```python
#There are duplicate records
df_no_duplicates = df.drop_duplicates(subset=['Innings Player','Innings Date'])
df = df_no_duplicates.copy()
```

```python
df
```

```python
#convert score to integer
df['Innings Runs Scored Num'] = pd.to_numeric(df['Innings Runs Scored Num'], errors='coerce')
df = df.dropna(subset=['Innings Runs Scored Num']).astype({'Innings Runs Scored Num' : 'int'})
```

```python
df
```

```python
df_sachin = df[df['Innings Player']=='SR Tendulkar']
df_sachin
```

```python
df_virat = df[df['Innings Player']=='V Kohli']
df_virat
```

```python
import scipy.stats as stats

shapiro_test = stats.shapiro(df_sachin["Innings Runs Scored Num"])
print("Shapiro_lik p-value: ", shapiro_test.pvalue)
```

```python
# Q-Q Plot
stats.probplot(df_sachin["Innings Runs Scored Num"], fit=stats.norm, plot=plt)
plt.title("Q-Q Plot of exam marks")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

```python
shapiro_test = stats.shapiro(df_virat["Innings Runs Scored Num"])
print("Shapiro_lik p-value: ", shapiro_test.pvalue)
# Q-Q Plot
stats.probplot(df_virat["Innings Runs Scored Num"], fit=stats.norm, plot=plt)
plt.title("Q-Q Plot of exam marks")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

```python
# Interpretation 
if shapiro_test.pvalue <0.05:
    print("The data likely does not follow a normal distribution.")
else:
    print("The data may be normally distributed, but the q-q plot can provide further insights.")
    
```

```python
t_statistic, p_value = stats.ttest_ind(df_sachin['Innings Runs Scored Num'], df_virat['Innings Runs Scored Num'])
print(t_statistic, p_value)
```

```python
#Conclusion
if p_value < 0.05:
    print(
        "Conclusion: Statistically significant difference between batting performance of SR Tendulkar and V Kohli means (p-value<0.05), so reject H0")
else:
    print(
        "Conclusion: No statistically significant difference between batting performance of SR Tendulkar and V Kohli (p-value >= 0.05), so do not reject H0")
```

```python
print(df_sachin["Innings Runs Scored Num"].mean(),df_virat["Innings Runs Scored Num"].mean())
```

```python
from scipy.stats import t
```

```python
deg_free = len(df_sachin) + len(df_virat) - 2
alpha = 0.05
critical_value = t.ppf(1 - alpha / 2, deg_free)
print(critical_value)
```

```python
if (t_statistic > critical_value):
    print("Reject H0, because t-statistics greater than critical")
else:
    print("Fail to Reject H0, because t-statistics Not greater than critical")
```
