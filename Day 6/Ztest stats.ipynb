```python
from statsmodels.stats.weightstats import ztest
```

```python
data =[88,89,90,92,97,999,95,89,101,105,108,112,96,112,114,109,109,102]

hypothesized_population_mean =150

```

```python
#put hypothesized mean value here

z_statistics ,p_value =ztest(data,value=hypothesized_population_mean)

#if the observed sample mean is less than the hypothesized population mean , the z_statistics qill be negative because of the observed measn is less that expected.Else it will poitive

print(f"Z-statistics: {z_statistics},p-value :{p_value}")
```

```python
#now visualize original data

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.stats import norm

#fit a normal distribution to the data
mu,std = norm.fit(data)

#plot the histogram 
plt.hist(data,bins=10,density=True,alpha=0.6,color='g')
#plot the pdf of fitted normal distribution
xmin ,xmax =plt.xlim()
x = np.linspace(xmin,xmax,100)
p=norm.pdf(x,mu,std)

plt.plot(x,p,'k',linewidth=20)
plt.xlabel("Iq Level")
plt.ylabel("Probability Density")

plt.title('Fit results:mu= % 2,std = % 2'%  (mu,std))
plt.show()
```

```python
#show hypothesized mean

ecdf =np.arange(1,len(data)+1)/len(data)

percentile_at_test_value =np.interp(hypothesized_population_mean,np.sort(data),ecdf)

plt.step(np.sort(data),ecdf,label='Empirical CDF',where='post')

plt.axvline(hypothesized_population_mean,color='red',linestyle='dashed',linewidth=2,label='Test value')


plt.xlabel('IQ levels')
plt.ylabel("Cumulative Probability")

plt.title("Empirical CDF of IQ levels")

plt.legend()
plt.show()
```
