```python
import pandas as pd
df1=pd.read_csv('D:\DBDA\Adavance Stats\pandas\Men ODI Player Innings Stats - 20th Century.csv')
df2=pd.read_csv('D:\DBDA\Adavance Stats\pandas\Men ODI Player Innings Stats - 21st Century.csv')
df3=pd.concat([df1,df2],axis=0)
df3
```

```python
srt = df3.loc[df3['Innings Player'] == 'SR Tendulkar']
srt
```

```python
df_no_duplicates = srt.drop_duplicates(subset=['Innings Player', 'Innings Date'])
srt = df_no_duplicates.copy()

srt['Innings Runs Scored Num'] = pd.to_numeric(srt['Innings Runs Scored Num'], errors = 'coerce')
srt = srt.dropna(subset=['Innings Runs Scored Num']).astype({'Innings Runs Scored Num' : 'int'})

mean_score = srt['Innings Runs Scored Num'].mean()
sd_score = srt['Innings Runs Scored Num'].std()
srt['z-score'] = (srt['Innings Runs Scored Num'] - mean_score) / sd_score
print('Innings Runs scored are total: ', srt['z-score'])

```

```python
srt['z_score'] = (srt['Innings Runs Scored Num'] - srt['Innings Runs Scored Num'].mean()) / srt['Innings Runs Scored Num'].std()
srt
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt


shapiro_test =stats.shapiro(srt['Innings Runs Scored Num'])
print("Shapiro-Wilk p-value:", shapiro_test.pvalue)

#Q-Q Plot
stats.probplot(srt['Innings Runs Scored Num'], fit=stats.norm, plot=plt)
plt.title("Q-Q Plot of Runs")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()

#Interpretation
if shapiro_test.pvalue < 0.05:
        print("THe data likely does not follow a normal distribution.")
else:
        print("The data may be normally distributed, but the q-q plot can provide further insights.")

# Analyze the q-q plot:
  # - If the points roughly fall along a straight line, it suggests normality.
# - Deviations from the line, especially at the tails, indicate non-normality.  
```
