```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("D:/DBDA/Adavance Stats/pandas/tips.csv")
df.head()
```

```python
df.shape
```

```python
ct =pd.crosstab(df['sex'],df['time'])
print(ct)
```

```python
ct_sum =pd.crosstab(df['sex'],df['time'],values=df['total_bill'],aggfunc='sum',margins=True)
print(ct_sum)
```

```python
ct_mean =pd.crosstab(df['sex'],df['time'],values=df['total_bill'],aggfunc='mean')
print(ct_mean)
```

```python
#create a crosstab table of SEX and TIME level with row and column margins 


ct_margins =pd.crosstab(df['sex'],df['time'],margins=True)
sns.heatmap(ct_margins,cmap='coolwarm',annot=True,fmt='d')
plt.show()
print(ct_margins)
```

```python
#create a crosstab table of sex and time  with normalization
#normalize = divide all the values by the sum of the values

ct_norm =pd.crosstab(df['sex'],df['time'],normalize=True,margins=True)
print(ct_norm)
```

```python
ct =pd.crosstab(df['sex'],df['smoker'])
print(ct)
```

```python

ct_margins =pd.crosstab(df['sex'],df['day'],margins=True)
sns.heatmap(ct_margins,cmap='coolwarm',annot=True,fmt='d')

print(ct_margins)
```

```python
ct_viz =pd.crosstab(df['sex'],df['time'])

ct_viz.plot(kind='bar',stacked=True)
plt.show()
print(ct_viz)
```

```python
ct_viz =pd.crosstab(df['sex'],df['day'])

ct_viz.plot(kind='bar',stacked=True)
plt.show()
print(ct_viz)
```
