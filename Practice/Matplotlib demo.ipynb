```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#seed sets values for generating random
np.random.seed(42) 

# linspace generates evenly spaced points in between 0 and 100
x = np.linspace(0,10,100)

y1 =np.random.normal(0,1,100)

y2= np.sin(x)+np.random.normal(0,0.2,100)

#generates 100 random values from a uniform distribution between 0 and 1
y3 = np.random.uniform(0,1,100)

```

```python
#creating various charts

plt.figure(figsize =(12,18))

#line plot

plt.subplot(3,3,1)
plt.plot(x,y1,label='Random Data') #plotting the line

plt.title('Line plot') #setting title

plt.legend() #displayng legend

#scatter plot

plt.subplot(3,3,2)
plt.scatter(x,y2,label='Sine wave with noise ',color='red') #plotting scatter plot
plt.title('Scatter Plot')
plt.legend()



#histogram

plt.subplot(3,3,3)
plt.hist(y3,bins=20,color='green',alpha=0.7)
plt.title("Histogram ")



#Bar plot
#compares different groups por categories
categories =['A','B','c']
values =[3,7,5]
plt.subplot(3,3,4)
plt.bar(categories,values,color='purple')   #creating a bar chart
plt.title("Bar Chart")



#Pie Chart
plt.subplot(3,3,5)
explode =(0,0.1,0) #explode the second slice by 0.1
plt.pie(values,labels=categories,autopct='%1.1f%%',startangle=90,explode=explode,colors=['red','blue','green']) 
plt.title("Pie Chart")


#Box Plot

#visualize the distribution and spread of dataset including outliers

plt.subplot(3,3,6)
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)  #creates a box plot
plt.title("Box plot")


#3D plot
#represents 3d relationship data
ax=plt.subplot(3,3,7,projection='3d')
z=np.linspace(0,1,100)
ax.plot(x,y1,z,label='3D Line plot')   #creating 3d line plot
ax.set_title('3D Line plot') #setting title
ax.legend()
#show the plots
plt.tight_layout()
plt.show()
```
