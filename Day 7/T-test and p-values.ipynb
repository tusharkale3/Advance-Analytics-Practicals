```python
import scipy.stats as stats
import matplotlib.pyplot as plt 
```

```python
#define data groups 
group_a =[85,88,90,92,95,87,89,91,94,86,88,92,90,93,85,89,91,87,92,90,88,93,86,89,91]
group_b=[78,82,80,85,88,81,83,79,84,87,80,82,85,78,81,83,79,82,85,80,78,81,83,79,84,87,80,82,85,78]

```

```python
#perform t -test

t_statistics,p_value=stats.ttest_ind(group_a,group_b)

print("T-Statistics ",t_statistics)
print("P-value",p_value)
```

```python
#conclusion based on p value

if p_value < 0.05:
    print("Conclusion : Statistically significant difference between Group A group B means (p-values <0.05), so reject H0")
else:
        print("Conclusion : No Statistically significant difference between Group A group B means (p-values >=0.05), so do not  reject H0")
```

```python
# QQ plot 

stats.probplot(group_a,dist='norm',plot=plt)
plt.title("QQ plot of Group A")
plt.show()

stats.probplot(group_b,dist='norm',plot=plt)
plt.title("QQ plot of Group B")
plt.show()
```

```python
# now scatter plot

plt.figure(figsize=(10,6))
plt.scatter(range(len(group_a)),group_a,color='blue',marker='o',label='Group A')
plt.scatter(range(len(group_b)),group_b,color='red',marker='s',label='Group B')
plt.xlabel("Data point index")
plt.ylabel("Score")
plt.title("COmparisioon of Group A and Group B scores")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#finding critical value 
from scipy.stats import t

df = len(group_a)+len(group_b) -2 #gor unequal varinace
alpha=0.05
critical_value =t.ppf(1-alpha/2,df)  #two tailed test
print(critical_value)
```

```python
if t_statistics>critical_value:
    print("Reject H0, because t_statistic > critical_value")
else:
    print("Fail to Reject H0, because t_statistic NOT > critical_value")
```
