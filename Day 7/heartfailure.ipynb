```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from scipy import stats 




```

```python
df = pd.read_csv("D:\DBDA\Adavance Stats\Day 7\heart_failure_clinical_records_dataset.csv")
df.head()
```

```python
sns.histplot(x = df['platelets'], kde = True);
plt.show()

print('mean = ', np.mean(df['platelets']))
print('median = ', np.median(df['platelets']))
print('std = ', np.std(df['platelets']))
```

```python
stats.probplot(df['platelets'], plot = plt);
plt.show()
```

```python
# print the skewness of each dataset
print(stats.skew(df['platelets']))
```

```python
# returns a numerical value representing the skewness of the dataset
print(stats.skewtest(df['platelets']))
```
