```python
import scipy.stats as stats

shapiro_test = stats.shapiro(df_new_sachin["Innings Runs Scored Num"])
print("Shapiro_lik p-value: ", shapiro_test.pvalue)
```

```python
# Q-Q Plot
stats.probplot(df_new_sachin["Innings Runs Scored Num"], fit=stats.norm, plot=plt)
plt.title("Q-Q Plot of exam marks")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

```python
# Interpretation 
if shapiro_test.pvalue <0.05:
    print("The data likely does not follow a normal distribution.")
else:
    print("The data may be normally distributed, but the q-q plot can provide further insights.")
    
```

```python
#finding critical value 
from scipy.stats import t


#perform t -test

t_statistics,p_value=stats.ttest_ind(df_SRT['Innings Runs Scored Num'],df_VK['Innings Runs Scored Num'])

print("T-Statistics ",t_statistics)
print("P-value",p_value)

srtvk = len(df_SRT['Innings Runs Scored Num'])+len(df_VK['Innings Runs Scored Num']) -2 #for unequal varinace
alpha=0.05
critical_value =t.ppf(1-alpha/2,srtvk)  #two tailed test
print(critical_value)
```

```python
if t_statistics>critical_value:
    print("Reject H0, because t_statistic > critical_value")
else:
    print("Fail to Reject H0, because t_statistic NOT > critical_value")
```
