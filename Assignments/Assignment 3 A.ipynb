```python
#changing '-' from column Margin to 'no result' 
df21['Margin'].fillna("no result",inplace=True)
```

```python
#concatenating given 2 dataframes in single
df = pd.concat([df20, df21], axis=0)
df
len(df)
```

```python
#checking null values count for df
df.isnull().sum()
```

```python
# creating new data frame df1 where we are eliminating N/R Can abn Tied from Results column
df1 =df[(df['Result'] !="N/R") & (df['Result'] !="Aban") & (df['Result'] !="Tied") & (df['Result'] !="Canc") & (df['Country'] =="India")]
df1
```

```python
#Renaming column Home?Away to HomeAway
df1.rename(columns={'Home/Away':"HomeAway"},inplace=True)
```

```python
#making a dataframe where we are selecting entries having country==India
df_ind = df1.query("Country =='India'")
df_ind
```

```python
# Probability of matches where Loaction is Home and Results are Won By India

p_a_and_b=len(df_ind.query("HomeAway=='Home' and Result=='Won'"))/len(df_ind)
p_a_and_b
```

```python
# Probability of matches where Loaction is Home  India
p_a=len(df_ind.query("HomeAway=='Home'"))/len(df_ind)
p_a

```

```python
# Probability of matches where Results are Won By India
print("Probability of matches where Results are Won By India")
p_b =len(df_ind.query("Result=='Won'"))/len(df_ind)
p_b
```

```python

# Probability of matches where Loaction is Home and Results were Won By India
p_a_given_b =(p_a_and_b)/p_b

p_a_given_b
```

```python
# Probability of matches where Results are Won and Loaction was Home By India
p_b_given_a =(p_a_and_b)/p_a

p_b_given_a
```

```python
#bayes thm
# Probability of matches where Results are Won and Loaction was Home By India
p_b_given_a_by_bayes =(p_a_given_b *p_b)/p_a
print("Probability of matches where Results are Won and Loaction was Home for country India",p_b_given_a_by_bayes)
```
